As the data has skewed distribution, therefore using Capping & Percentile Method for outlier removal

```python
def percentile_capping(attribute):
  # Computing 2th, 98th percentiles and replacing the outliers
  tenth_percentile = np.percentile(data[attribute], 2)
  ninetieth_percentile = np.percentile(data[attribute], 98)
  data[attribute] = np.where(data[attribute]<tenth_percentile, tenth_percentile, data[attribute])
  data[attribute] = np.where(data[attribute]>ninetieth_percentile, ninetieth_percentile, data[attribute])
  return data[attribute]
```

```python
data['lead_time'] = percentile_capping('lead_time')
data['stays_in_weekend_nights'] = percentile_capping('stays_in_weekend_nights')
data['stays_in_week_nights'] = percentile_capping('stays_in_week_nights')
data['adults'] = percentile_capping('adults')
data['children'] = percentile_capping('children')
data['previous_bookings_not_canceled'] = percentile_capping('previous_bookings_not_canceled')
data['booking_changes'] = percentile_capping('booking_changes')
data['adr'] = percentile_capping('adr')
data['required_car_parking_spaces'] = percentile_capping('required_car_parking_spaces')
data['total_of_special_requests'] = percentile_capping('total_of_special_requests')
```

```python
sns.histplot(data= data, x = 'adr', bins = 50)
```

* Average Daily Rate(ADR) measures the average rental revenue earned for an occupied room per day.
* Most of the ADR is around 100, but there is sudden spike near 0 ADR stating zero occupancy

```python
count, bins_count = np.histogram(data['adr'], bins=10)
pdf = count / sum(count)

#numpy np.cumsum for calculating the CDF
cdf = np.cumsum(pdf)

plt.plot(bins_count[1:], pdf, color="red", label="adr PDF")
plt.plot(bins_count[1:], cdf, label="adr CDF")
plt.legend()
plt.title("ADR CDF & PDF")
```

```python
# Lead time is most of the times 0
sns.histplot(data= data, x = 'lead_time', bins = 50)
```

* Lead time measures the number of days that elapsed between the entering date of the booking into the PMS and the arrival date.
* Most of the people are booking very close to arrival date
* Very few people book hotel very early before arrival date

```python
count, bins_count = np.histogram(data['lead_time'], bins=10)
pdf = count / sum(count)

#numpy np.cumsum for calculating the CDF
cdf = np.cumsum(pdf)

plt.plot(bins_count[1:], pdf, color="red", label="lead_time PDF")
plt.plot(bins_count[1:], cdf, label="lead_time CDF")
plt.legend()
```
